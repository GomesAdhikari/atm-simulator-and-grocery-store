```python
# ATM module ..........

x = input("Enter Bank Account Number : ")
Balance = 909090909
if len(x)!=9:
    print("Enter a valid Account number!")
else:
    y = input("Enter your Pin : ")
    if len(y)<8:
        print("Wrong pin!")
    else:
        print("Welcome User!👨")
        print("1. Check Bank Balance\n2. Change your pin\n3. Withdraw money\n4. Deposit money ")
        z = input("Enter your choice : ")
        opt = ["1","2","3","4"]
        if z not in opt:
            print("Enter a valid choice!")
        else:
            if z=="1":
                print("Your Bank Balance is : ",Balance)
            if z=="2":
                c = input("Enter old pin : ")
                if c!=y:
                    print("Wrong Pin! Please retry.")
                else:
                    if c == y:
                        i = input("Enter new pin : ")
                        if len(i)<8:
                            print("Pin must be at least 8 characters")
                        else:
                            print("Pin changed Successfully!")
            if z=="3":
                money = range(0,10000)
                for amount in money:
                    amount = int(input("Enter Amount : "))
                    if amount<5000 :
                        print("Amount withdrawn sucessfully take your cash ⭐")
                        print("Your updated balance is : ",Balance - amount)
                        x = input("Do you want to receipt? :  ")
                        if x==("yes","Yes"):
                            print("Receipt printed")

                        break
                    else:
                         print("Cannot withdraw more than 5000 !")
            if z=="4":
                print("Reminder : old currency denominations are not accepted if you want to deposit old denominations\n please visitit to the nearest SBI branch.")
                z = [500,100,200]
                notes_count = 0
                notes_try = 3
                while notes_try>notes_count:
                    notes = int(input("Enter currency denomination : "))
                    notes_count +=1
                    if notes in z:
                        amount = int(input("Enter amount : "))
                        if amount%10!=0 or amount<500:
                            print("Deposit amount should be divisible by 10 and ust be more than 500.")
                        else:
                            print("Amount deposited Current balance : ",Balance + amount)
                else:
                    print("Warning : If you will not report to SBI within 24 Hours\nThe system will report your KYC information to Mukhani police station")
```
